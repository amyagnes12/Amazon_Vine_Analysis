```python
# Dependency
import pandas as pd

# File to load
vine_review = "Resources/vine_table.csv"

# Read data and store into a pandas df
vine_table_df = pd.read_csv(vine_review)
vine_table_df.head(10)
```

```python
total_votes_df = vine_table_df[vine_table_df["total_votes"] >= 20]
total_votes_df.head(10)
```

```python
filtered_votes_df = total_votes_df[total_votes_df["helpful_votes"]/total_votes_df["total_votes"] >= 0.5]
filtered_votes_df.head(10)
```

```python
paid_vine_df = filtered_votes_df[filtered_votes_df["vine"] == "Y"]
paid_vine_df.head(10)
```

```python
unpaid_vine_df = filtered_votes_df[filtered_votes_df["vine"] == "N"]
unpaid_vine_df.head(10)
```

```python
# Total number of reviews for paid (Vine program)
paid_num_reviews = len(paid_vine_df.index)
paid_num_reviews
```

```python
# Total number of reviews for unpaid
unpaid_num_reviews = len(unpaid_vine_df.index)
unpaid_num_reviews
```

```python
# Total number of 5-star reviews for paid (Vine program)
paid_five_star = paid_vine_df[paid_vine_df["star_rating"] == 5.0].count()["review_id"]
paid_five_star
```

```python
# Total number of 5-star reviews for unpaid
unpaid_five_star = unpaid_vine_df[unpaid_vine_df["star_rating"] == 5.0].count()["review_id"]
unpaid_five_star
```

```python
# percentage of 5-star reviews for paid (Vine program)
paid_five_star/paid_num_reviews
```

```python
# percentage of 5-star reviews for unpaid
unpaid_five_star/unpaid_num_reviews
```
